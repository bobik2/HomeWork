```python
import numpy as np
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
A = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
w = np.linalg.solve(A, b)
X=np.arange(1,16)
Y1 = X * w[1] + w[0]
F=f(X)

import matplotlib.pyplot as plt
plt.plot(X,Y1)
plt.plot(X,F)
```

```python
A2 = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b2 = np.array([f(1), f(8), f(15)])
w2 = np.linalg.solve(A2, b2)
X=np.arange(1,16)
Y2 = X*X * w2[2] + X * w2[1] + w2[0]
plt.plot(X,Y2)
plt.plot(X,F)
```

```python
a = np.array([1,4,10,15])
A3 = np.array([a**0, a**1, a**2, a**3]).T
b3 = np.array([f(1), f(4), f(10), f(15)])
w3 = np.linalg.solve(A3, b3)
X=np.arange(1,16)
Y3 = X*X*X * w3[3] + X*X * w3[2] + X * w3[1] + w3[0]
plt.plot(X,Y3)
plt.plot(X,F)
print(w3)
```
